# Linear Regression Using Gradient Descent 

```python
# Step-1 
m = 0
c = 0
lr = 0.0001
n = len(x)

for i in range(500):

    # Step-2 
    # yhat --> means Prediciton 
    yhat=(m*x)+c
    Dm = (-2/n)*sum(x*(y-yhat))
    Dc = (-2/n)*sum(y-yhat)

    # Step-3
    #update the parameters
    m = m-(lr*Dm)
    c = c-(lr*Dc)

    # Print korelam --> ami within M and C er moddhe korteci (Loop er bhitor e korlam)
    print(m,c)
```

### To solve the ebove issue : 

```python
# Initialize variables
m = 0
c = 0
lr = 0.0001
x = np.array(x, dtype=float)  # Convert x to a numpy array with float type
y = np.array(y, dtype=float)  # Convert y to a numpy array with float type
n = len(x)

for i in range(500):

    # Step-2 
    # yhat --> means Prediction 
    yhat = (m * x) + c
    
    # Calculate derivatives
    Dm = (-2 / n) * np.sum(x * (y - yhat))
    Dc = (-2 / n) * np.sum(y - yhat)

    # Step-3
    # Update the parameters
    m = m - (lr * Dm)
    c = c - (lr * Dc)

    # Print values of m and c after each iteration
    print(m, c)
```
